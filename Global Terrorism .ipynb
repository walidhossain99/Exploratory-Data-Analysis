# checking null values (yellow portion indicates null)

```python
plt.figure(figsize=(16,6)) #plotting the figure
sns.heatmap(dataset.isnull(),cmap='viridis',cbar=True) #plotting the dataset into a heatmap
```

Lots of Features with a huge number of Null Values.So, it will be better to just get rid of those features

# Selecting the columns we need

```python
dataset = dataset[['eventid','iyear', 'imonth', 'country_txt', 'region_txt','natlty1_txt', 'provstate', 'city', 'latitude', 'longitude', 'success', 'attacktype1_txt', 'targtype1_txt', 'natlty1_txt', 'gname', 'weaptype1_txt', 'weapsubtype1_txt', 'nkill', 'nwound']]

#selecting only the columns we need
```

```python
dataset.rename(columns={'eventid':'Event ID','iyear':'Year','imonth':'Month','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'Latitude','longitude':'Longitude','attacktype1_txt':'Attack Type','targtype1_txt':'Target Type','gname':'Group','natlty1_txt': 'Nationality','motive':'Motive','weaptype1_txt':'Weapon Used', 'weapsubtype1_txt': 'Weapon subtype','nkill':'Killed','nwound':'Wounded'},inplace = True)

# renaming the columns we selected
```

```python
dataset.head() #printing our first 5 selected columns 
```

### checking null values once again for selected columns

```python
plt.figure(figsize=(16,6))
sns.heatmap(dataset.isnull(), cmap='viridis',cbar=False) 
```

looks pretty neat than the previous one. Still there are some little null values in the dataset.

### Filling the nulls in the data

```python
dataset['Weapon subtype']=dataset['Weapon subtype'].fillna('Unknown') # filling the null values with a string
dataset['Nationality'] = dataset['Nationality'].fillna('No Record')  # filling the null values with a string
```

```python
dataset.head()
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(dataset.isnull(), cmap='viridis',cbar=False)
```

### Dropping the nulls

```python
print('No. of rows before droppig nulls : {}'.format(dataset['Event ID'].count()))
dataset.dropna(inplace=True)
print('No. of rows after dropping nulls : {}'.format(dataset['Event ID'].count()))
```

```python
dataset['Casualties'] = dataset['Killed'] + dataset['Wounded'] #adding the killed and wounded into one feature
```

```python
plt.figure(figsize =(16,6))
sns.heatmap(dataset.isnull(),cmap='viridis', cbar = False)
```

```python
dataset['Year'].unique() #looking at the number of unique years 
```

# Looking at the correlation between the features

```python
plt.figure(figsize=(20,5))
sns.heatmap(dataset.corr(),annot = True)
```

# No. of attacks worldwide (1970 - 2017)

```python
sns.set_style('darkgrid')
ls = dataset['Year'].value_counts().sort_index()
plt.figure(figsize = (28,5))
Most_att = sns.barplot(ls.index, ls.values,alpha = 0.8)
plt.title('Most Attacks Worldwide')
plt.xlabel('Years')
plt.ylabel('Counts')
Most_att.set_xticklabels(labels = ls.index, fontsize = 13)
plt.show()
```

As we can see that from 2011 the Attacks started to spike high and at 2014, we can see the highest number of attacks

# Most affected Regions

```python
dataset[dataset['Region'].isin(dataset['Region'].value_counts()[:20].index)][['Casualties','Region']].groupby('Region').sum().reset_index().sort_values('Casualties',ascending=True).merge(dataset[dataset['Region'].isin(dataset['Region'].value_counts(ascending = True)[:20].index)]['Region'].value_counts(ascending = False).to_frame().reset_index().rename(columns={'index':'Region','Region':'Attacks'}),on='Region').plot.barh(x='Region',y=['Casualties','Attacks'], figsize =(16,6),stacked=True)
```

Here, we can see the Middle East and North Africa are the regions to get most affected by terrorist attacks followed by South Asia and other regions.

#  Most affected countries

```python
dataset[dataset['Country'].isin(dataset['Country'].value_counts()[:20].index)][['Casualties','Country']].groupby('Country').sum().reset_index().sort_values('Casualties', ascending = False).merge(dataset[dataset['Country'].isin(dataset['Country'].value_counts()[:20].index)]['Country'].value_counts().to_frame().reset_index().rename(columns = {'index':'Country','Country':'Attacks'}),on='Country').plot.bar(x='Country', y=['Attacks','Casualties'],figsize=(16,6),stacked=True)
plt.title('Most affected Countries', fontsize=15)
plt.xlabel('Country', fontsize = 15)
```

Among the Middle East Region, Irag seems to have been the most affected country.

# Most affected States

```python
sns.set_style('darkgrid')
ls = dataset['State'].value_counts()
ls = ls[:20]
plt.figure(figsize=(20,5))
aff_st = sns.barplot(ls.index,ls.values, alpha=0.8)
plt.title('Most affected States',fontsize = 15)
plt.xlabel('States',fontsize = 15)
plt.ylabel('Attacks',fontsize = 15)
aff_st.set_xticklabels(labels = ls.index, rotation = 90,fontsize =15)
plt.show()
```

# Most active terrorist groups

```python
ls = dataset['Group'].value_counts().sort_values(ascending = False)
ls = ls[:5]
plt.figure(figsize=(22,6))
most_act = sns.barplot(ls.index, ls.values, alpha = 0.8)
plt.title("Top 5 active groups")
plt.xlabel('Group names')
plt.ylabel('Activity')
most_act.set_xticklabels(labels = ls.index, rotation = 0,fontsize =14)
plt.show()
```

The plot represting the Top 5 active groups participating the attacks that have taken place

# Most Attack type Used 

```python
ls = dataset['Attack Type'].value_counts().sort_values(ascending = False)
ls = ls[:5]
plt.figure(figsize=(22,6))
most_att = sns.barplot(ls.index, ls.values, alpha = 0.8)
plt.title("Top 5 Attacks Methonds Used")
plt.xlabel('Attack Type')
plt.ylabel('Activity')
most_att.set_xticklabels(labels = ls.index, rotation = 0,fontsize =14)
plt.show()
```

The Top 5 methods used to govern these attacks

# Top 10 Targets of the Terrorists

```python
ls = dataset['Target Type'].value_counts().sort_values(ascending = False)
ls = ls[:10]
plt.figure(figsize=(22,6))
most_tar = sns.barplot(ls.index, ls.values, alpha = 0.8)
plt.title("Top 10 Targets",fontsize = 13)
plt.xlabel('Targets',fontsize = 15)
plt.ylabel('Activity',fontsize = 13)
most_tar.set_xticklabels(labels = ls.index, rotation = 50,fontsize =13)
plt.show()
```

Here we can that the Private Citizens and Properties were the top targets of the terrorist groups.
